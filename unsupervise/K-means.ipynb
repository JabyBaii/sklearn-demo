```python
import pandas as pd
```

```python
# 1、获取数据
order_products = pd.read_csv("../data_big/order_products__prior.csv")
products = pd.read_csv("../data_big/products.csv")
orders = pd.read_csv("../data_big/orders.csv")
aisles = pd.read_csv("../data_big/aisles.csv")
```

```python
order_products.head()
```

```python
products.head()
```

```python
orders.head()
```

```python
aisles.head()
```

```python
# 2、合并表
# order_products__prior.csv：订单与商品信息

# 字段：order_id, product_id, add_to_cart_order, reordered
# products.csv：商品信息
# 字段：product_id, product_name, aisle_id, department_id
# orders.csv：用户的订单信息
# 字段：order_id,user_id,eval_set,order_number,….
# aisles.csv：商品所属具体物品类别
# 字段： aisle_id, aisle

# 合并aisles和products aisle和product_id
tab1 = pd.merge(aisles, products, on=["aisle_id", "aisle_id"])
```

```python
tab2 = pd.merge(tab1, order_products, on=["product_id", "product_id"])
```

```python
tab3 = pd.merge(tab2, orders, on=["order_id", "order_id"])
```

```python
tab3.head()
```

```python
# 3、找到user_id和aisle之间的关系
table = pd.crosstab(tab3["user_id"], tab3["aisle"])
```

```python
table.shape
```

```python
data = table[:10000]
data.shape
```

```python
# 4、PCA降维
from sklearn.decomposition import PCA
```

```python
# 1）实例化一个转换器类
transfer = PCA(n_components=0.95)

# 2）调用fit_transform
data_new = transfer.fit_transform(data)
```

```python
data_new.shape
```

```python
data_new
```

```python
# 将降维后的数据进行预估器流程
## 看结果
## 模型评估

from sklearn.cluster import KMeans
```

```python
estimator = KMeans(n_clusters=3)
# 没有目标值,这里只传特征值
estimator.fit(data_new)
```

```python
y_predict = estimator.predict(data_new)
```

```python
# 被分成3类
y_predict[:300]
```

```python
# 模型性能评估-轮廓系数
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(data_new, y_predict)
```
