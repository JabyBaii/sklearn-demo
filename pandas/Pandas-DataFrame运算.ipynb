```python
import pandas as pd
data = pd.read_csv("../data_wj/stock_day.csv")
data = data.drop(["v_ma5","v_ma10","v_ma20"],axis=1)
```

```python
data
```

# 1.算数运算

```python
data['open'].add(1)  # 此列都加1
```

```python
data['high'].sum()  # 求此列的和
```

# 2.逻辑运算

```python
data['open'] > 23.4
```

```python
data[data['open'] > 23.4]  # 逻辑判断的结果作为筛选条件
```

```python
data[(data['open'] > 23.4) & (data['open'] < 24)].head()  # 多个逻辑判断
```

```python
data.query("open>23.4 & open<24").head()  # 使用逻辑运算函数得到同样的结果
```

```python
data[data["open"].isin([23.53, 23.70])]  # 指定条件进行筛选
```

# 累积统计

```python
data = data.sort_index()
```

```python
stock_rise = data["p_change"]
stock_rise
```

```python
stock_rise.cumsum()
```

## 用图形展示

```python
import matplotlib.pyplot as plt
stock_rise.cumsum().plot()
plt.show()
```
