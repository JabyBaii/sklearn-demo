```python
import pandas as pd
```

```python
data = pd.read_csv("../data_wj/stock_day.csv")
```

```python
data = data.drop(["v_ma5","v_ma10","v_ma20"],axis=1)
```

```python
data
```

# 1.索引取值（先列后行）

```python
data["open"]["2018-02-23"]
```

# 2.使用loc

```python
data.loc["2018-02-27":"2018-02-22", "high"]
```

```python
data.loc["2018-02-27":"2018-02-22", "high":"low"]
```

# 3.使用iloc

```python
data.iloc[:3, :5]  # 获取前3行,前5列
```

# 4. 选取指定的列

```python
data.index[0:4]
```

```python
data.loc[data.index[0:4], ['open','high','close','volume']]
```

```python
data.columns.get_indexer(['open','high','close','volume'])
```

```python
data.iloc[0:4, data.columns.get_indexer(['open','high','close','volume'])]
```
