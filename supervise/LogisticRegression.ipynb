```python
import pandas as pd 
import numpy as np
```

```python
# 列索引
"""
   #  Attribute                     Domain
   -- -----------------------------------------
   1. Sample code number            id number
   2. Clump Thickness               1 - 10
   3. Uniformity of Cell Size       1 - 10
   4. Uniformity of Cell Shape      1 - 10
   5. Marginal Adhesion             1 - 10
   6. Single Epithelial Cell Size   1 - 10
   7. Bare Nuclei                   1 - 10
   8. Bland Chromatin               1 - 10
   9. Normal Nucleoli               1 - 10
  10. Mitoses                       1 - 10
  11. Class:                        (2 for benign, 4 for malignant)
"""
column_name = ['Sample code number', 'Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape', 'Marginal Adhesion',
              'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin', 'Normal Nucleoli', 'Mitoses', 'Class']
data = pd.read_csv("../data_ml/breast+cancer+wisconsin+original/breast-cancer-wisconsin.data", names=column_name)
data.head(60)
```

```python
data = data.replace(to_replace="?", value=np.nan)
# 检查是否有缺失值
data.isnull().any()
```

```python
# 删除缺失样本
data.dropna(inplace=True)
data.isnull().any()
```

```python
data
```

```python
# 划分数据集
from sklearn.model_selection import train_test_split
```

```python
# 刷选特征值和目标值
x = data.iloc[:, 1:-1]
y = data["Class"]
```

```python
x.head()
```

```python
y.head()
```

```python
# 划分数据集
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
x_train
```

```python
# 特征工程-标准化
from sklearn.preprocessing import StandardScaler
```

```python
transfer = StandardScaler()
# 标准化训练集和测试集
x_train = transfer.fit_transform(x_train)
x_test = transfer.transform(x_test)
```

```python
# 标准化后,均值为0,标准差为1
x_test
```

```python
# 逻辑回归预估器
from sklearn.linear_model import LogisticRegression
```

```python
estimator = LogisticRegression()
estimator.fit(x_train, y_train)
```

```python
# 逻辑回归模型的参数: 回归系数和偏置
estimator.coef_
```

```python
estimator.intercept_
```

```python
# 模型评估
y_predict = estimator.predict(x_test)
print("y_predict:\n", y_predict)
print("通过模型预测的值是否等于测试集的真实值\n", y_test == y_predict)

score = estimator.score(x_test, y_test)
print("模型对测试集的准确率为：\n", score)
```

```python
# 精确率/ 召回率/ F1-score
from sklearn.metrics import classification_report
report = classification_report(y_test, y_predict, labels=[2, 4], target_names=["良性", "恶性"])
```

```python
print(report)
```

```python
# 处理样本不均衡问题 ROC AUC
# 将y_test转化为0（反例）,1（正例）表示
y_true = np.where(y_test > 3, 1, 0)
```

```python
y_true
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_true, y_predict)
```
